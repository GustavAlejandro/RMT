**Rotational Invariant Estimators (RIE)**

So far our random matrices of particular interest are of the form
\begin{align*}
    \mathbf{E} = \mathbf{C} + \mathbf{X} \text{   or  } \mathbf{E} = \mathbf{C}^{1/2}\mathbf{W}\mathbf{C}^{1/2}
\end{align*}
where $\mathbf{X}$ and $\mathbf{W}$ represent some noise, where $\mathbf{X}$ might be a Wigner matrix and $\mathbf{W}$ a Wishart matrix, whereas $\mathbf{C}$ is the uncorrupted matrix that we would like to measure. Two main question arise:

1.   How similar are sample eigenvectors $\left[\mathbf{v}_{i}\right]_{i\in(1,N)}$ of $\mathbf{E}$ and the true ones $\left[\mathbf{u}_{i}\right]_{i\in(1,N)}$ of $\mathbf{C}$?
2.   What information can we learn by observing two independent realizations $(\mathbf{X} , \mathbf{X}' \text{or} \mathbf{W},\mathbf{W}')$ that remain correlated through the same $\mathbf{C}$?

We define $\mathbf{\Xi}(\mathbf{E})$ an estimator of $\mathbf{C}$ given $\mathbf{E}$, this estimator is rotationally invariant iif

\begin{align*}
    \mathbf{\Xi}(\mathbf{OEO}^{\intercal}) = \mathbf{O}\mathbf{\Xi}(\mathbf{E})\mathbf{O}^{\intercal}
\end{align*}

for any orthogonal matrix $\mathbf{O}$. If $\mathbf{E}$ is rotated by some $\mathbf{O}$, then $\mathbf{C}$ must be rotated in the same fashion. Naively, this is because we have no prior assumption on the eigenvectors of $\mathbf{C}$, so the only special directions in which $\mathbf{C}$ can point are those singled out by $\mathbf{E}$ itself. The estimators $    \mathbf{\Xi}(\mathbf{OEO}^{\intercal}) = \mathbf{O}\mathbf{\Xi}(\mathbf{E})\mathbf{O}^{\intercal}$ are called rotationally invariant estimators (RIE).

$\mathbf{\Xi}(\mathbf{E})$ can be diagonalized in the same basis as $\mathbf{E}$, up to a fixed rotation matrix $\mathbf{\Omega}$ but there is no natural guess for it, except the identity matrix $\mathbf{1}$. Hence $\mathbf{\Xi}(\mathbf{E})$ has the same eigenvectors $(\mathbf{v}_{i})$ as those of $\mathbf{E}$, then:

\begin{align*}
    \mathbf{\Xi}(\mathbf{E}) = \sum_{i = 1}^{N}\xi_{i}\mathbf{v}_{i}\mathbf{v}^{\intercal}_{i}
\end{align*}

where $\xi_{k}$ is a function of all empirical eigenvalues $\left[\lambda_{k}\right]_{k\in(1,N)}$.








To find the optimal RIE we define the Loss function as:

\begin{align*}
    \mathcal{L}(\mathbf{\Xi}(\mathbf{E}),\mathbf{C}) = || \mathbf{\Xi}(\mathbf{E}) -\mathbf{C}   ||_{\text{F}}^{2}
\end{align*}

Where $|| \cdot  ||_{\text{F}}^{2}$ is the Frobenius norm defined by:

\begin{align*}
    || A  ||_{\text{F}}^{2} = \text{Tr}(A^{2})
\end{align*}


\begin{align*}
    \mathcal{L}(\mathbf{\Xi}(\mathbf{E}),\mathbf{C})=\text{Tr}\left(\mathbf{\Xi}(\mathbf{E}) - \mathbf{C}\right)^{2} = \sum_{i=1}^{N}\mathbf{v}^{\intercal}_{i}\left(\mathbf{\Xi}(\mathbf{E}) - \mathbf{C}\right)^{2}\mathbf{v}_{i} = \sum_{i = 1}^{N}\left(\xi_{i}^{2}-2\xi_{i}\mathbf{v}^{\intercal}_{i}\mathbf{C}\mathbf{v}_{i}+\mathbf{v}^{\intercal}_{i}\mathbf{C}^{2}\mathbf{v}_{i}\right)
\end{align*}

Then we take the derivative of $\mathcal{L}$ with respect to $\xi_{k}$ equals to zero

\begin{align*}
    \frac{\partial \mathcal{L}}{\partial \xi_{k}} = 2\xi_{l} -2\mathbf{v}^{\intercal}_{l}\mathbf{C}\mathbf{v}_{l} = 0
\end{align*}

which implies

\begin{align*}
    \xi_{k} = \mathbf{v}^{\intercal}_{k}\mathbf{C}\mathbf{v}_{k}
\end{align*}

Well, this is ridiculous! We don't have a clue of $\mathbf{C}$ and we find an equation for the $\xi$ that is impossible to compute because we don't know $\mathbf{C}$!

This is called the 'oracle' estimator because knowledge we do not have. Long story short, we are not screwed whatsoever. There is another route we can take.

After many complicated computations and sophisticated arguments it can be proven that
\begin{align*}
    \xi_{k} = \frac{1}{\pi\rho_{\mathbf{E}}(\lambda_{k})} \lim_{\eta \rightarrow 0^{+}}Z(z_{k})g_{\mathbf{E}}(z_{k})-Y(z_{k})\text{ ,     } z_{k}=\lambda_{k}-i\eta
\end{align*}

with
\begin{align}
  Z(z) = \frac{z}{z-q+qzg_{\mathbf{E}}(z)}
\end{align}

and
\begin{align*}
    Y(z) = \begin{cases}1 & \text{additive case}\\ Z(z)/z & \text{multiplicative case}\end{cases}
\end{align*}

**Additive case**

\begin{align*}
    \mathbf{E} = \mathbf{C} + \mathbf{X}
\end{align*}

For a general noise matrix $\mathbf{X}$, $  Z(z) = z - R_{\mathbf{X}}(g_{\mathbf{E}}(z))$

then
\begin{align*}
    \xi(\lambda) = \lambda - \frac{\lim_{\eta \rightarrow 0^{+}}\text{Im}R_{\mathbf{X}}(g_{\mathbf{E}}(z))g_{\mathbf{E}}(z)}{\lim_{\eta \rightarrow 0^{+}}\text{Im}g_{\mathbf{E}}(z)}, z = \lambda - i\eta
\end{align*}

if there is no noise $\mathbf{X} = 0$ and hence $R_{\mathbf{X}} = 0$ and we get the trivial case $\xi(\lambda) = \lambda$

 A case to consider is when $\mathbf{X}$ and $\mathbf{C}$ are Wigner matrices, then $\mathbf{E}$ is also a Wigner matrix. If $\sigma^{2}_{x}$ and $\sigma^{2}_{c}$ are the variances of $\mathbf{X}$ and $\mathbf{C}$ respectively then $\mathbf{E}$ has variance $\sigma^{2} = \sigma^{2}_{x} + \sigma^{2}_{c}$, with With $R_{\mathbf{X}} = x\sigma^{2}_{x}$

(I will compute just the results, Latex makes me sick :C)
In this case we have:

\begin{align*}
    \xi(\lambda) = \lambda - \frac{\lambda\sigma^{2}_{x}}{\sigma^{2}} = r\lambda
\end{align*}

Where $r = \sigma^{2}_{c}/\sigma^{2}$

Then

\begin{align*}
    \mathbf{\Xi}(\mathbf{E})  = \sum_{i = 1}^{N}\xi_{i}\mathbf{v}_{i}\mathbf{v}^{\intercal}_{i} = r\sum_{i = 1}^{N}\lambda_{i}\mathbf{v}_{i}\mathbf{v}^{\intercal}_{i} = r\mathbf{E}
\end{align*}

\begin{align*}
    \mathbf{\Xi}(\mathbf{E})_{jk} = r\mathbf{E}_{jk}
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python

def semicircle(s2 = 1, n = 100):
  x = np.linspace(-2*np.sqrt(s2) + (1/n), 2*np.sqrt(s2) - (1/n), n)
  r = np.sqrt(4*s2 - x**2)/(2*np.pi*s2)
  return x, r

def arcsine_law(limit, n = 50):
  x = np.linspace(-limit+(1/n),limit-(1/n), n)
  rho = 1/(np.pi*np.sqrt(limit**2 - x**2))
  return x, rho

class matrix(object):
  def __init__(self, N, T = None, s = 1,  mtype = None):
    self.N = N
    self.T = T

    if mtype == None or 'wigner':
      H = np.random.normal(0,s, size =(self.N,self.N))
      self.W = (H+H.T)/(np.sqrt(2*self.N))

    if mtype == 'wishart':
      self.q = (N/T)
      self.p = (self.q)/(1-self.q)
      H = np.random.normal(0,s, size = (self.N, self.T))
      self.W = (1/self.T)*H@H.T

  #return self.W

  def orthogonal(self):

    uw, vw = np.linalg.eig(self.W)
    self.M = vw@np.diag(np.sign(uw))@vw.T

    return self.M

  def inverse_w(self):

    self.IW = (1-self.q)*np.linalg.inv( self.W )

    return self.IW
```

The free product of two Wishart matrices $\mathbf{E} = (\mathbf{W}_{q0})^{\frac{1}{2}}\mathbf{W}_{q}(\mathbf{W}_{q0})^{\frac{1}{2}}$.

Where $\mathbf{W} = \frac{1}{T}\mathbf{HH}^{\intercal}$ and $q = N/T$

In this case

\begin{align*}
    \xi(\lambda) = \lambda \frac{\lim_{\eta \rightarrow 0^{+}}\text{Im}S_{\mathbf{W}}(t_{\mathbf{E}}(z))t_{\mathbf{E}}(z)}{\lim_{\eta \rightarrow 0^{+}}\text{Im}t_{\mathbf{E}}(z)}\text{  , }  z = \lambda - i\eta
\end{align*}

then

\begin{align*}
    \xi(\lambda) = \frac{\lambda}{|1+qt_{\mathbf{E}}(\lambda - i\eta)|^{2}}
\end{align*}

or

\begin{align*}
    \xi(\lambda) = \frac{\lambda}{|1-q+q\lambda g_{\mathbf{E}}(\lambda - i\eta)|^{2}}
\end{align*}


For a Wishart matrix with parameter $q = N/T$ it can be proven that the S-Transform is given by:
\begin{align}
    S_{W_{q}}(t) = \frac{1}{1 + qt}
\end{align}

Then the product is:

\begin{align*}
    S_{E}(t) = S_{W_{q}}(t)S_{w_{q0}}(t) =  \frac{1}{1 + qt}  \frac{1}{1 + q_{0}t}
\end{align*}

finding $\mathrm{t}_{E}(\zeta)$

\begin{align*}
    S_{E}(t) = \frac{t+1}{t\zeta_{E}(t)} \Rightarrow t\zeta_{E}(t)= \frac{t+1}{S_{E}(t)}
\end{align*}

then

\begin{align*}
    qq_{0}t^{3} + (q + q_{0} + qq_{0})t^{2} + (q + q_{0} + 1 - \zeta)t + 1 = 0
\end{align*}

```python
def polyroots(zeta,eta = 0,q = 1, q0 = 1):
  x3 = q*q0
  x2 = q + q0 + q*q0
  x1 = q + q0 + 1
  x0 = 1
  root = []
  for zz in zeta:
    r = np.roots([x3, x2, x1 - (zz - 1*1j*eta), x0])
    root.append(r)
  #return np.ravel(root)
  return np.array(root)
```
